```python
df = pd.read_csv('weather_data.csv', parse_dates=['Day'], index_col='Day')
df #preview
```

```python
#generic filling
df.fillna(0)
```

```python
#filling with pre filled values
df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
```

```python
# filling automatically with ffill
df.fillna(method='ffill')
```

```python
# filling automatically with bfill
t = df.fillna(method='bfill')
t

%matplotlib inline
t.plot()
```

```python
# filling automatically with ffill plus limit
df.fillna(method='ffill', limit=1)
```

```python
# filling by interpolation 
df.interpolate()
```

```python
# filling by interpolation  with time method
gp = df.interpolate(method='time')
```

```python
gp.describe()
```

```python
#dropping empty cells
df.dropna()
```

```python
#dropping empty rows cells
df.dropna(how='all')
```

```python
#dropping empty cells
#adding limits
df.dropna(thresh=1) #at least one value for it to be preserved
```

```python
#dropping empty cells
#adding limits
df.dropna(thresh=2) #at least two values for it to be preserved
```

```python
import numpy as np
#handling data replace
df = pd.read_excel('weather.xlsx')
df
```

```python
#replacing empty data

df.replace(-9999, np.NaN)
```

```python
#replacing different empty data

df.replace([-9999, 'No Event'], np.NaN)
```

```python
#replacing empty data with different data

df.replace({
    'temperature': -9999,
    'windspeed': -9999,
    'event': 'No Event'
}, np.NaN)
```

```python
#replacing empty data with different data

df.replace({
    -9999: np.NaN,
    'No Event': 'Sunny'
})
```
